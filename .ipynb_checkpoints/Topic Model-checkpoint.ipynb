```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('input/tweets.csv')
df.head()
```

```python
def clean_tweet(twdf):
    junk = re.compile("al|RT|\n|&.*?;|http[s](?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+)*")
    clean_tweets = [junk.sub(" ", t) for t in twdf.tweets]
    return clean_tweets
```

```python
tweets = clean_tweet(df)
tweets[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_df=.5)
tfidfvec = vec.fit_transform(tweets)
```

```python
words = vec.get_feature_names()
```

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components=10).fit(tfidfvec)
for i, topic in enumerate(nmf.components_):
    print("Topic ", i, ": ")
    print(" ".join([words[i] for i in topic.argsort()[:-10 - 1:-1]]))
```

```python
df['topic'] = np.argmax(nmf.transform(vec.transform(tweets)), axis=1)
top10users = df[df.username.isin(df.username.value_counts()[:10].keys().tolist())]
pd.crosstab(top10users.username, top10users.topic).plot.bar(stacked=True, figsize=(16, 10), colormap="coolwarm")
```
